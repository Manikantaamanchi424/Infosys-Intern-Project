```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("titanic.csv")
```

```python
#1.Overall survival rate
overall_survival_rate = df["Survived"].mean() * 100
print(f"Overall Survival Rate: {overall_survival_rate:.2f}%")
```

```python
#2.How does the survival rate vary by passenger class (Pclass)?
survival_by_class = df.groupby("Pclass")["Survived"].mean() * 100
print("\nSurvival Rate by Class:\n", survival_by_class)
```

```python
#3.What is the average age of passengers who survived vs. those who did not survive?
avg_age_survival = df.groupby("Survived")["Age"].mean()
print("\nAverage Age (0 = Not Survived, 1 = Survived):\n", avg_age_survival)
```

```python
#4.How many passengers embarked from each port (Embarked), and what are their survival rates?
embarked_counts = df["Embarked"].value_counts()
embarked_survival = df.groupby("Embarked")["Survived"].mean() * 100
print("\nPassenger Counts by Embarked:\n", embarked_counts)
print("\nSurvival Rate by Embarked:\n", embarked_survival)
```

```python
#5.What is the distribution of fare prices paid by passengers, and does fare impact survival chances?
print("\nFare Distribution Summary:\n", df["Fare"].describe())
fare_survival = df.groupby(pd.qcut(df["Fare"], 4))["Survived"].mean() * 100
print("\nSurvival Rate by Fare Quartiles:\n", fare_survival)
```

```python
#6.Plot the count of passengers by passenger class (Pclass).
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Pclass", palette="viridis")
plt.title("Passenger Count by Class")
plt.show()
```

```python
#7.Visualize the survival rate with respect to gender using a bar chart.
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="Sex", y="Survived", palette="Set2")
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
#8.Show the age distribution of passengers using a histogram or KDE plot.
plt.figure(figsize=(6,4))
sns.histplot(df["Age"].dropna(), bins=30, kde=True, color="blue")
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
#9.Create a boxplot comparing fares across different passenger classes.
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Pclass", y="Fare", palette="pastel")
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
#10.Visualize survival count by embarkation port (Embarked) with a stacked bar chart.
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Embarked", hue="Survived", palette="Set1")
plt.title("Survival Count by Embarkation Port")
plt.show()
```

```python
#11.Display a heatmap of the correlation matrix of numerical features in the dataset.
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```
